```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv
data = pd.read_csv('Housing.csv')
print(data)
```

```python
data.describe()
```

```python
data.plot(x='area', y='bedrooms', kind='line', title='trend of numeric variable over time', marker='o', linestyle='-', color='b')
plt.show()
```

```python
x = 'area' 
y = 'price'  

# Create a scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(data[x], data[y], alpha=0.5)

# Customize the plot
plt.title("Scatter Plot of Housing Prices")
plt.xlabel(x)
plt.ylabel(y)

# Display the plot
plt.show()
```

```python
import seaborn as sns
numeric_vars = ['area', 'price', 'bedrooms', 'bathrooms', 'stories', 'parking']
corr_matrix = data[numeric_vars].corr()

plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', square=True)

# Customize the plot
plt.title("Correlation Matrix Heatmap")
plt.xticks(rotation=60)
plt.yticks(rotation=30)

# Display the plot
plt.show()
```
